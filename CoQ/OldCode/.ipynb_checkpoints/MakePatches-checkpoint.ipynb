```python
import sys
sys.path.insert(0, "/Users/aimachine/PyImage/utils")
import numpy as np

import importlib
import matplotlib.pyplot as plt
import os
import glob
import skimage
from skimage import measure
from skimage import filters
from Normalize import normalizeFloat, normalizeMinMax, Path, save_tiff_imagej_compatible
from skimage.filters import threshold_otsu
from tifffile import imread
```

```python
basedir = '/Users/aimachine/MakePatches/'
targetdir =  '/Volumes/TRANSCEND/TestPatches/'
```

```python
Path = os.path.join(basedir, '*.tif')
BigT = []
Names = []
filesRaw = glob.glob(Path)
maxtime = 0
axes = 'TYX'
for fname in filesRaw:
      x = imread(fname)
     
      
      BigT.append(x)
   
     
      Names.append(fname)
      
    
BigT.sort
Names.sort

CommonName = 'Patch'
print('Total number of input images = ', len(BigT))
print('Image size = ', BigT[0].shape)
```

```python
ShapeX = int(75)
ShapeY = int(75)
ShapeT = int(7)



for i in range(0, len(BigT)):
  img = BigT[i]
  timeextend = img.shape[0]
  sizeY = img.shape[2]  
  sizeX = img.shape[1]  
  while timeextend%ShapeT!=0:
        
        timeextend = timeextend + 1
  while sizeX%ShapeX!=0:
        sizeX = sizeX + 1
        
  while sizeY%ShapeY!=0:
        sizeY = sizeY + 1
  extendimage = np.zeros([timeextend,sizeX, sizeY])
  extendimage[:img.shape[0], :img.shape[1], :img.shape[2]]  = img 
  print(extendimage.shape)
  plt.imshow(extendimage[5])
  maxjumpX = int((extendimage.shape[1])/ShapeX)  
  maxjumpY = int((extendimage.shape[2])/ShapeY)
  maxid = int(extendimage.shape[0]/ShapeT)
  spacejumpX = 0
  spacejumpY = 120  
  for idx in range(0,maxid):

         
          smallimg = extendimage[idx:idx + ShapeT,spacejumpX:spacejumpX + ShapeX,spacejumpY:spacejumpY + ShapeY]
          save_tiff_imagej_compatible((targetdir + CommonName + str(idx) + str(spacejumpX) + str(spacejumpY)+ '.tif') , smallimg, axes)
```
